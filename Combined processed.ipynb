```python
import pandas as pd

# Reading csv files 
google_store = pd.read_csv("google_store_processed.csv")
apple_store = pd.read_csv("apple_store_processed.csv")

# Combining the csv files
data = pd.concat([google_store, apple_store], ignore_index = False)

data
```

```python
data.info()
```

```python
data = data.drop_duplicates(subset=['Review'])
data
```

```python
data.to_csv('combined_review_data.csv' , index = False)
```
